## Clean up the data
Did you notice anything unusual about the "house" amounts?

How about the handset price and income?

```python
df_clean = df[(df['house'] > 0) & (df['income'] > 0) & (df['handset_price'] < 1000)]
df_clean.describe()
```

# Add Derived Attribute

Add a column with the income in K (thousands)

```python
df_clean['income_k'] = (df_clean['income']/1000).round(2)
df_clean.head()
```

## Summarize the Data
 What is average house price at the two levels of leave?

 Conclusion?  Stayers have substantially more expensive homes.

```python
df_clean[['house','leave']].groupby('leave').mean().round(2)
```

## Additional Data Preparation

```python
# prompt: add at least two additional steps to clean or filter dataset

#remove rows with missing values
df_clean = df_clean.dropna()
#remove rows with duplicate values
df_clean = df_clean.drop_duplicates()
```

## Data Visualization

 Assignment Tasks

```python
# prompt: Add at least two charts to show a summary of any major trends

#histogram of house prices
mpl.pyplot.hist(df_clean['house'])
mpl.pyplot.show()
#scatter plot of income vs house price
mpl.pyplot.scatter(df_clean['income_k'], df_clean['house'], s=.5)
mpl.pyplot.show()
```
